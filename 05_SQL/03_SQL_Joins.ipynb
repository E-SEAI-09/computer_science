```python
%%sql
drop table if exists authors cascade;

CREATE TABLE IF NOT EXISTS authors (
  id serial primary key,
  name varchar(255),
  birth_year int
);

insert into authors (name, birth_year)
	values
   ('Eric Meyer', 1960),
	 ('Estelle Weyl', 1964),
	 ('Astrid Lindgren', 1907),
   ('J.R.R. Tolkien', 1892);
```

```python
%%sql
select * from authors;
```

foreign key (author_id) references authors(id)


```python
%%sql
drop table if exists books cascade; 

create table if not exists books (
  id serial primary key,
  title varchar(255),
  release_year int,
  author_id int,
  foreign key (author_id) references authors(id)
);

insert into books (title, release_year, author_id)
  values 
  ('CSS: The Definite Guide', 2023, 1),
  ('Ronja Räubertochter', 1981, 3),
  ('The Lord of the Rings', 1954, 4),
  ('The Silmarillion', 1977, 4),
  ('Book without author', 2026, null);
```

```python
%%sql
select * from books;
```

```python
%%sql
select a.name as "Author's name", b.title
  from authors as a
  join books as b -- same as inner join
    on a.id = b.author_id;
```

```python
%%sql
select a.name as "Author's name", b.title
  from authors as a
  left join books as b -- all from left table + matching books
    on a.id = b.author_id;
```

```python
%%sql
select a.name as "Author's name", b.title
  from authors as a
  full join books as b -- all from both tables
    on a.id = b.author_id;
```

```python
%%sql
select a.name as "Author's name", b.title
  from authors as a
  right join books as b -- matching from authors table and all books
    on a.id = b.author_id
```

```python
%%sql
select a.name as "Author's name", b.title, a.birth_year
  from authors as a
  join books as b
    on a.id = b.author_id
  where a.birth_year < 1950;
```

```python
%%sql
drop table if exists authors cascade;
drop table if exists books;
```
