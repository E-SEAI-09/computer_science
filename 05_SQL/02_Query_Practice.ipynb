1. List the titles of all content released in 2008


```python
%%sql 

SELECT title FROM netflix_shows
  where release_year = 2008
  limit 5;
```

2. List the titles of all content with a release year on or after 2018, in alphabetical order.


```python
%%sql 

SELECT title FROM netflix_shows
  where release_year >= 2018
  order by title
  limit 10;
```

3. Determine the number of movies with a maturity rating of PG-13


```python
%%sql 

SELECT count(*) FROM netflix_shows
  where type = 'Movie' and rating = 'PG-13';
```

4. List the maturity ratings, and determine the number of content with each rating


```python
%%sql 

select rating, count(*) from netflix_shows
  group by rating
  order by count(*) desc;
```

5. List the titles and release years of all Barbie movies, in chronological order.
   1. You can assume all Barbie movies will start with “Barbie”, and that if a movie starts with “Barbie” it is a Barbie movie.


```python
%%sql 

SELECT title, release_year FROM netflix_shows
  where title ilike 'bArBiE%' and type = 'Movie'
  order by release_year;
```

6. List the names of all people who starred in a movie in which Kevin Bacon also starred.


```python
%%sql 

SELECT cast_members FROM netflix_shows
  where cast_members like '%Kevin Bacon%';
```

```python
%%sql 

SELECT distinct unnest(string_to_array(cast_members, ', ')) as actor FROM netflix_shows
  where cast_members like '%Kevin Bacon%';
```

```python
%%sql
SELECT DISTINCT TRIM(actor) AS actor_name
FROM netflix_shows,
  unnest(string_to_array(cast_members, ',')) AS actor
  WHERE type = 'Movie'
  AND cast_members LIKE '%Kevin Bacon%'
  AND TRIM(actor) <> 'Kevin Bacon';
```

7. List the title, release year, rating and description of TV shows where:
   1. Matt LeBlanc appeared
   2. Gerard Butler starred
   3. Benedict Cumberbatch appeared


```python
%%sql 
SELECT title, release_year, rating, description
  FROM netflix_shows
  WHERE type = 'TV Show'
  AND 
    (cast_members ilike '%Matt LeBlanc%' 
    or  cast_members ilike '%Gerard Butler%' 
    or cast_members ilike '%Benedict Cumberbatch%');
```

8. List the title, release year, rating and description of Movies where:
   1. Robert Pattinson appeared
   2. Quentin Tarantino directed


```python
%%sql 

SELECT title, release_year, rating, director, description FROM netflix_shows
  where type = 'Movie'
    and
      (cast_members ilike '%Robert Pattinson%'
      and director = 'Werner Herzog');
```

9. List the types, and determine the average release year of each type


```python
%%sql 

SELECT type, avg(release_year)::int FROM netflix_shows
  group by type;
```
